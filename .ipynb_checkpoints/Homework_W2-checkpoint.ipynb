```python
import pandas as pd
import os
```

```python
ue = pd.read_json('data/students_UE.json')
ue.head()
import matplotlib.pyplot as plt
plt.hist(ue['Grade UE'], bins = 10)
plt.rcParams["figure.figsize"] = (5,5)
plt.ylabel('Count')
plt.xlabel('Grade')
plt.title("Grade UE")
plt.show()
```

```python
vl = pd.read_json('data/students_VL.json')
vl.head()
plt.hist(vl['Grade VL'], bins = 10)
plt.rcParams["figure.figsize"] = (7,5)
plt.ylabel('Count')
plt.xlabel('Grade')
plt.title("Grade VL")
plt.show()
```

```python
from statistics import mean
print('Average grade for UE class is ',mean(ue['Grade UE']))
print('Average grade for VL class is ',mean(vl['Grade VL']))
```

```python
combined = pd.merge(ue, vl[['Matrikelnummer','Grade VL']], on = 'Matrikelnummer')
combined.head()
```

```python
combined['UE - VL']=combined['Grade UE']-combined['Grade VL']
```

```python
combined = combined.sort_values(by='UE - VL')
#plt.hist(combined['Grade UE'],bins=25)
#plt.hist(combined['Grade VL'],bins=25)
combined['UE - VL'].plot.hist(bins=25)
plt.rcParams["figure.figsize"] = (5,5)
plt.xlabel('UE - VL')
plt.title("Difference between UE and VL class grade")
plt.show()
```

```python
attendance = pd.read_csv('data/students_attendance.csv')
attendance.head()
```

```python
attendance_check = pd.merge(combined, attendance[['Matrikelnummer','attendence']], on = 'Matrikelnummer')
attendance_check['AVG.Grade'] = (attendance_check['Grade UE']+attendance_check['Grade VL'])/2
onl_studs = attendance_check.query('attendence == "online"')
incl_studs = attendance_check.query('attendence=="in_class"')
print('AVG.Grade of in class student ',
    mean(incl_studs['AVG.Grade']))
print('AVG.Grade of online student ',
    mean(onl_studs['AVG.Grade']))
```

### Revenues

First, load the data from CSV file

```python
revenues = pd.read_csv('data/revenues.csv')
revenues.head()
```

To be compatible with time series, need to rotate (transpose) the dataframe

```python
data = revenues.transpose() #transpose dataframe -> index also transpose -> create new index, del wrong index
data = data[1:] #del wrong index
data = data.reset_index() #create new index
data.columns = ['Week','F1','F2','F3','F4'] #set column names
data.head()

plt.rcParams["figure.figsize"] = (15,5) #width, height of data frame

rev_plot = data.plot.line(title="Revenues of 4 branches in 100 weeks") 
#set other variable name so that if change happen, it won't affect older code
rev_plot.set_xlabel('Week')
rev_plot.set_ylabel('Revenue')
plt.show()
```

```python
data.describe()
rev_mean = data.mean()
rev_std = data.std()
rev_plot_std = rev_mean.plot.bar(
    title="Mean and deviation of 4 branches",
    yerr=rev_std
)
rev_plot_std.set_xlabel("Branch").set_color('blue')
rev_plot_std.set_ylabel("Revenue").set_color('red')
```

```python
import statistics
line1 = pd.read_csv('data/production_line1.csv')
arr1 =[]
for index, row in line1.iterrows():
  arr1.append(float(row.item()))
print('Mean produce time of line1: ',mean(arr1))
print('Standard deviation of produce time of line1: ',statistics.stdev(arr1))
for index, row in line1.iterrows():
  arr1.append(float(row.item()))
```

```python
line2 = pd.read_csv('data/production_line2.csv')
arr2 =[]
for index, row in line2.iterrows():
  arr2.append(float(row.item()))
print('Mean produce time of line2: ',mean(arr2))
print('Standard deviation of produce time of line2: ',statistics.stdev(arr2))
for index, row in line2.iterrows():
  arr2.append(float(row.item()))
```

```python
line3 = pd.read_csv('data/production_line3.csv')
arr3 =[]
for index, row in line3.iterrows():
  arr3.append(float(row.item()))
print('Mean produce time of line3: ',mean(arr3))
print('Standard deviation of produce time of line3: ',statistics.stdev(arr3))
for index, row in line3.iterrows():
  arr3.append(float(row.item()))
```

```python
# import numpy as np
combined_chart=[arr1,arr2,arr3]
plt.rcParams["figure.figsize"] = (7,5)
fig1, ax1 = plt.subplots()
ax1.set_title('Box Plot')
ax1.boxplot(combined_chart)
```

```python
plt.rcParams["figure.figsize"] = (7,5)
fig2, ax2 = plt.subplots()
ax2.set_title('Violin Plot')
ax2.violinplot(combined_chart)
```
